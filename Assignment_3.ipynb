```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the daiab data
data = pd.read_csv('diab.csv')
```

```python

# Set a seed for reproducibility
np.random.seed(45)
```

```python
# Take a random sample of 35 observations
sample = data.sample(n=35)
```

```python
# Calculate the mean and highest Glucose values for the sample
sample_mean_glucose = sample['Glucose'].mean()
sample_highest_glucose = sample['Glucose'].max()
```

```python
# Calculate the population mean and highest Glucose values
population_mean_glucose = data['Glucose'].mean()
population_highest_glucose = data['Glucose'].max()
```

```python
# Print the statistics
print("Sample Mean Glucose:", sample_mean_glucose)
print("Sample Highest Glucose:", sample_highest_glucose)
print("\nPopulation Mean Glucose:", population_mean_glucose)
print("Population Highest Glucose:", population_highest_glucose)
```

```python
# Create a bar chart for comparison
labels = ['Sample Mean', 'Sample Highest', 'Population Mean', 'Population Highest']
values = [sample_mean_glucose, sample_highest_glucose, population_mean_glucose, population_highest_glucose]
```

```python
plt.bar(labels, values, color=['yellow', 'orange', 'purple', 'magenta'])
plt.title('Comparison of Glucose Statistics between Sample and Population')
plt.ylabel('Glucose Values')
plt.show()
```

```python
# Find the 98th percentile of BMI for the sample and population
sample_percentile_bmi = np.percentile(sample['BMI'], 98)
population_percentile_bmi = np.percentile(data['BMI'], 98)
```

```python
# Print the percentiles
print("Sample 98th Percentile BMI:", sample_percentile_bmi)
print("Population 98th Percentile BMI:", population_percentile_bmi)
```

```python
# Create a bar chart for comparison
labels = ['Sample 98th Percentile', 'Population 98th Percentile']
values = [sample_percentile_bmi, population_percentile_bmi]
```

```python

plt.bar(labels, values, color=['purple', 'magenta'])
plt.title('Comparison of 98th Percentile BMI between Sample and Population')
plt.ylabel('BMI Values')
plt.show()
```

```python
# Set a seed for reproducibility
np.random.seed(45)
```

```python
# Create 500 bootstrap samples (with replacement) of 150 observations each
bootstrap_samples = [data.sample(n=150, replace=True) for _ in range(500)]
```

```python
# Calculate the average mean, standard deviation, and percentile for BloodPressure in each bootstrap sample
bootstrap_means = [sample['BloodPressure'].mean() for sample in bootstrap_samples]
bootstrap_std_devs = [sample['BloodPressure'].std() for sample in bootstrap_samples]
bootstrap_percentiles = [np.percentile(sample['BloodPressure'], 95) for sample in bootstrap_samples]
```

```python
# Calculate the population mean, standard deviation, and percentile for BloodPressure
population_mean_bp = data['BloodPressure'].mean()
population_std_dev_bp = data['BloodPressure'].std()
population_percentile_bp = np.percentile(data['BloodPressure'], 95)
```

```python
# Create histograms for comparison
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.hist(bootstrap_means, bins=30, color='blue', alpha=0.7, label='Bootstrap Means')
plt.axvline(population_mean_bp, color='red', linestyle='dashed', linewidth=2, label='Population Mean')
plt.title('Comparison of Means for BloodPressure')
plt.legend()
```

```python
plt.subplot(1, 3, 2)
plt.hist(bootstrap_std_devs, bins=30, color='green', alpha=0.7, label='Bootstrap Std Devs')
plt.axvline(population_std_dev_bp, color='red', linestyle='dashed', linewidth=2, label='Population Std Dev')
plt.title('Comparison of Standard Deviations for BloodPressure')
plt.legend()
```

```python
plt.subplot(1, 3, 3)
plt.hist(bootstrap_percentiles, bins=30, color='orange', alpha=0.7, label='Bootstrap Percentiles')
plt.axvline(population_percentile_bp, color='red', linestyle='dashed', linewidth=2, label='Population 95th Percentile')
plt.title('Comparison of 95th Percentiles for BloodPressure')
plt.legend()
```
